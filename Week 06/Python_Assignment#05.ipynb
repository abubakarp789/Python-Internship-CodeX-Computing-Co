## Part (A)

```python
import pandas as pd
```

#### 1. Load the dataset and display the first 10 rows.

```python
data = pd.read_csv("Datasets/StudentsPerformance.csv")
```

```python
data.head(10)
```

#### 2. Check for missing values in the dataset.

```python
data.isnull().sum()
```

#### 3. Rename column names to remove spaces and lowercase them and show the output by displaying the columns name after changings.

```python
data = data.rename(columns=lambda x: x.strip().replace(' ', ''))
```

```python
data.columns
```

#### 4. Create a new column average_score as the mean of math, reading, and writing scores. 

```python
data['average_score'] = data[['mathscore', 'readingscore', 'writingscore']].mean(axis=1)
data['average_score']
```

#### 5. A: 90+, B: 80–89, C: 70–79, D: 60–69, F: below 60 (based on average_score), create a column for grades as per given logic.

```python
data['grade'] = data['average_score'].apply(lambda x: 'A' if x >= 90 else 'B' if x >= 80 else 'C' if x >= 70 else 'D' if x >= 60 else 'F')
data['grade']
```

#### 6. Display details of students who completed the test preparation course.

```python
completed_test_students = data[data['testpreparationcourse'] == 'completed']
completed_test_students
```

#### 7. Find all female students who scored above 85 in math.

```python
female_students = data[(data['mathscore'] >= 85) & (data['gender'] == 'female')]
female_students
```

#### 8. Group students based on parental level of education and calculate their average scores.

```python
grouped_students = data.groupby('parentallevelofeducation')['average_score'].mean()
grouped_students
```

#### 9. Group by gender and find the maximum average score. 

```python
max_average_score_by_gender = data.groupby('gender')['average_score'].max()
max_average_score_by_gender
```

#### 10. Show top 5 students with the highest average_score.

```python
top_5_students = data.nlargest(5, 'average_score')
top_5_students
```

#### 11. Create a second DataFrame with student_id and sports_participation, merge it with the original using inner join (you can add data from yourself in new dataframe) 

```python
data['student_id'] = range(1, 1001)

sports_participation_df = pd.DataFrame({
    'student_id': range(1, 1001),
    'sports_participation': ['yes' if i % 2 == 0 else 'no' for i in range(1, 1001)]
})

merged_df = pd.merge(data, sports_participation_df, on='student_id', how='inner')

merged_df.head()
```

#### 5. If missing values are less than 5% of data, drop the rows. otherwise, fill missing values using the mean of the column.

```python
missing_pct = housing_data.isnull().sum() / len(housing_data) * 100

for column in missing_pct.index:
    if missing_pct[column] < 5:
        housing = housing_data.dropna(subset=[column])

housing.fillna(method='bfill', inplace=True)
```

```python
housing.isnull().sum()
```
